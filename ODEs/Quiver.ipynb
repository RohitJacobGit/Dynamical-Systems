```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import math

def model(y,t):
    k = 0.3
    dydt = -k * y
    return dydt

# initial condition
y0 = 5

# time points
t = np.linspace(0,20)

# solve ODE
y = odeint(model,y0,t)

# plot results
plt.plot(t,y)
plt.xlabel('time')
plt.ylabel('y(t)')
plt.show()
```

```python
import sympy as sym
sym.init_printing()
t, lam = sym.symbols('t lambda')
x = sym.Function('y')(t)
dxdt = x.diff(t)
expr = sym.Eq(dxdt, -lam*x)
expr
```

```python

def f(y, t):
    """this is the rhs of the ODE to integrate, i.e. dy/dt=f(y,t)"""
    return np.sin(y)#-2 * y * t

y0 = 1             # initial value
a = 0              # integration limits for t
b = 2

t = np.arange(a, b, 0.01)  # values of t for
                          # which we require
                          # the solution y(t)
y = odeint(f, y0, t)  # actual computation of y(t)

import pylab          # plotting of results
pylab.plot(t, y)
pylab.xlabel('t'); pylab.ylabel('y(t)')
```

```python
def model(x,t,k):
    return -k * math.sin(x)

t = np.linspace(0,10)

x0=5

k=0.1
x1 = odeint(model,x0,t,args=(k,))
k=0.2
x2 = odeint(model,x0,t,args=(k,))
k=0.5
x3 = odeint(model,x0,t,args=(k,))

plt.plot(t,x1,'r-',linewidth=2,label='k=0.1')
plt.plot(t,x2,'b--',linewidth=2,label='k=0.2')
plt.plot(t,x3,'g:',linewidth=2,label='k=0.5')
plt.xlabel('time')
plt.ylabel('x(t)')
plt.legend()
plt.show()
```
